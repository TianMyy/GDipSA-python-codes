```python
import pandas as pd
import numpy as np
import string
import nltk
from nltk.corpus import stopwords
```

```python
file = open('space_invaders.txt', encoding='utf-8')
doc = file.read()
file.close()
```

```python
stop_words = set(stopwords.words('english'))

nltk.download('punkt')

porter = nltk.stem.PorterStemmer()
sentences = nltk.tokenize.sent_tokenize(doc)
print(sentences)
```

```python
# treat each sentence as a document
docs=[]
punc=str.maketrans('','',string.punctuation)
for ele in sentences:
    ele_nopunc = ele.translate(punc)
    words_stemmed = [porter.stem(w) for w in ele_nopunc.lower().split()]
    docs += [' '.join(words_stemmed)]

print(docs)
```

```python
for i in range(0,len(docs)):
    print('doc' + str(i) + '=',docs[i])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vec = TfidfVectorizer(analyzer='word',stop_words=stop_words)
tfidf_wm = tfidf_vec.fit_transform(docs).toarray()
print(tfidf_wm)

#生成一个矩阵，每个doc里面如果出现了该term就会有数值，没出现的数值就是0
```

```python
df_index = ['doc'+str(i) for i in range(len(docs))]
print(df_index)
#列出所有doc，成为接下来要组建的dataframe的行名
```

```python
df_columns = tfidf_vec.get_feature_names()
print(df_columns)
#列出所有unique的term,成为接下来要组建的dataframe的列名
```

```python
tfidf_df = pd.DataFrame(data=tfidf_wm,index=df_index,columns=df_columns)
tfidf_df

#构建dataframe
```

```python
# axis=1代表跨列，按行运算，axis=0代表跨行，按列运算
tfidf_sum_by_docs = np.sum(tfidf_df, axis=1) #此处将每个doc的数值分别求和
nwords_by_docs = np.sum(tfidf_df != 0, axis=1) #此处确认每个doc中有几个term
print(tfidf_sum_by_docs)
print(nwords_by_docs)
```

```python
nwords_by_docs = nwords_by_docs.sum(axis=0)/len(nwords_by_docs)
series = tfidf_sum_by_docs / nwords_by_docs
print(nwords_by_docs)
print(series)
```

```python
#将series排序，按照降序出现，便于我们选择前十个doc
sorted_series = series.sort_values(ascending=False)
top_series = sorted_series.head(10)

print("\n")

# remove the prefix 'doc' from indexes such as
# [doc10, doc24, doc11, ..., doc22]
num_only = [int(x[3:]) for x in top_series.index]
print(num_only)

for i in sorted(num_only):
    print(sentences[i] + "\n")
```
