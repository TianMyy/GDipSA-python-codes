### Load Necessary Libraries# 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# Resize Graph(让图变大了 dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(5,3), dpi=100)

## Line 1

# Kayword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, linestyle='--', marker='.', markersize=8, markeredgecolor='blue')

# Shorthand notation(可以有简单写法)
#fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2(第二根红线)

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')
# plot reminder of graph
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title(specify font parameters with fontdict(标题名称字体设置)
plt.title('Our first graph',fontdict={'fontname':'Comic Sans MS','fontsize':20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X,Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```
