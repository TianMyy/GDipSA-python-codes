```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# --1--
df = pd.read_csv('D:\SA-NUS\Python\Python ML 1 (Yuen Kwan)\Codes (ML)\Workshop\OmniPower.csv')
df
```

```python
# --2--
from sklearn import linear_model
from sklearn.model_selection import train_test_split

advert = df[['Price','Promotion']]
sales = df.Sales

x_train, x_test, y_train, y_test = train_test_split(advert,sales, random_state = 1)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)

from sklearn.linear_model import LinearRegression
linReg = LinearRegression()
linReg.fit(x_train, y_train)
print(linReg.intercept_)
print(linReg.coef_)
```

```python
# --3--
y_hat = linReg.predict(x_test)
print(x_test)
print(y_test)
print(y_hat)

ypredictions = linReg([[50,200]])
```

```python
# --4--
from sklearn.metrics import mean_squared_error,r2_score
print(mean_squared_error(y_test,y_hat))
print(r2_score(y_test,y_hat))
```
